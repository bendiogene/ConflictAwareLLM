## The Impact of Learning Rate Ratios on Catastrophic Forgetting: A Counterintuitive Discovery
### The way old knowledge was learned matters for how it is kept or not

The above experiments reveal a surprising relationship between learning rates and knowledge retention. Two key observations emerge:

1. Higher relative learning rates for new knowledge drive a clear trade-off:
   When learning new facts with a learning rate double that of the original old facts learning phase (0.001 vs 0.0005), we observe significantly faster acquisition of new knowledge, but at the cost of increased deterioration of old knowledge. This higher forgetting occurs even when employing targeted strategies that explicitly avoid neurons previously dedicated to old knowledge storage.

2. Matched learning rates (even if they are high) show unexpected resilience:
   Surprisingly, when both old and new knowledge are acquired using the same higher learning rate (0.001), the network demonstrates better retention of old information while maintaining the fast acquisition of new knowledge. This challenges the intuition that faster learning of new information necessarily leads to more forgetting.

This finding is particularly counterintuitive because even when employing sophisticated targeting strategies to protect old knowledge (avoiding "stubborn neurons"), the relative magnitude of learning rates still plays a crucial role in knowledge retention. The stability of encoded information appears to depend not just on which neurons encode it, but also on how it was encoded in the first place.

These results suggest that matching learning rates across learning phases might be as important as managing which neurons encode new information, opening new perspectives or warranting further investigation.
