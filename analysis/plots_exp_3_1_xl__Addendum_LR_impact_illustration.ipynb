The following script was used to generate the experiments that are visualized below below:

```bash
#!/bin/bash

# List of config files
configs=(
    "configs/gpt2-xl_experiment3_1_20000_1000.yml"
    "configs/gpt2-xl_experiment3_1_20000_1000_10x_lr.yml"
    "configs/gpt2-xl_experiment3_1_2000_1000.yml"
    "configs/gpt2-xl_experiment3_1_2000_1000_10x_lr.yml"
    "configs/gpt2-xl_experiment3_1_20000_100.yml"
    "configs/gpt2-xl_experiment3_1_20000_1000_10_small_lr.yml"
    "configs/gpt2-xl_experiment3_1_2000_100.yml"
    "configs/gpt2-xl_experiment3_1_2000_1000_10_small_lr.yml"
)

# Iterate over each config file and run the command
for config in "${configs[@]}"
do
    accelerate launch experiments_scripts/exp_3_1_xl.py --config "$config"
done

```

# 1. GPT2 XL in the exact same LR as GPT2 small (from 2k to 20k): 

## Takeaway: 
- Not enough to learn new knowledge (especially for 10 and 100 facts)
- With 1000 facts, some amount of knowledge is ingested, suggested that ease of knowledge ingestion likely depends on the facts at hand (with more facts, we might be more likely to find "suitable" facts)
- Even when we don't ingest enough new facts, the impact on old knowledge is quite surprising.

````python
strategies=['busy_A', 'free_A', 'spec_B', 'busy_B', 'rnd']
experiment_name = 'experiment3_1_xl'
filepath = '../experiments/gpt2-xl/experiment_3_1/'

filename = 'experiment_3_1_2000_10.json'
loc_old_1 = 'lower left'
bbox_old_1 = (0, 0)
loc_new_1 = 'upper left'
bbox_new_1 = (0, 1)
loc_gen_1 = 'upper left'
bbox_gen_1 = (0, 1)
y_lim_old_1_1 = 0
y_lim_old_2_1 = 1.02

display(Markdown('## 1.1. 10 facts (old => new => generalization)'))
plot_editing_old_new_general_noLoRa(experiment_name, filename, filepath, strategies, 
                             loc_old_1, bbox_old_1, loc_new_1, bbox_new_1, loc_gen_1, bbox_gen_1,
                             y_lim_old_1_1, y_lim_old_2_1)

# Experiment 2 (N_SAMPLES = 100)
filename = 'experiment_3_1_20240719_094011_2000_100.json'
loc_old_2 = 'lower left'
bbox_old_2 = (0, 0)
loc_new_2 = 'upper right'
bbox_new_2 = (1, 1)
loc_gen_1 = 'upper right'
bbox_gen_1 = (1, 1)
y_lim_old_1_2 = 0
y_lim_old_2_2 = 1.02
display(Markdown('## 1.1. 100 facts (old => new => generalization)'))
plot_editing_old_new_general_noLoRa(experiment_name, filename, filepath, strategies, 
                             loc_old_2, bbox_old_2, loc_new_2, bbox_new_2, loc_gen_1, bbox_gen_1,
                             y_lim_old_1_2, y_lim_old_2_2)

# Experiment 3 (N_SAMPLES = 1000)
filename = 'experiment_3_1_20240719_000646_2000_1000.json'
loc_old_3 = 'upper right'
bbox_old_3 = (1, 1)
loc_new_3 = 'upper right'
bbox_new_3 = (1, 1)
loc_gen_1 = 'upper right'
bbox_gen_1 = (1, 1)
y_lim_old_1_3 = 0
y_lim_old_2_3 = 1.02


display(Markdown('## 1.1. 1000 facts (old => new => generalization)'))
plot_editing_old_new_general_noLoRa(experiment_name, filename, filepath, strategies, 
                             loc_old_3, bbox_old_3, loc_new_3, bbox_new_3, loc_gen_1, bbox_gen_1,
                             y_lim_old_1_3, y_lim_old_2_3)
````
