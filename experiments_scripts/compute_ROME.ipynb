```python
import pandas as pd
from scipy.stats import hmean

# Results obtained by running evaluate_ROME on multiple seeds and fact sizes
data = {
    'Samples': [10, 10, 10, 10, 10, 100, 100, 100, 100, 100, 1000, 1000, 1000, 1000, 1000],
    'Seed': [46, 16, 5, 114, 1220, 46, 16, 5, 114, 1220, 46, 16, 5, 114, 1220],
    'Old (Locality)': [0.952, 0.745, 0.908, 0.948, 0.903, 0.435, 0.470, 0.554, 0.258, 0.437, 0.0261, 0.1906, 0.1808, 0.1756, 0.187],
    'New (Reliability)': [0.200, 0.300, 0.500, 0.000, 0.200, 0.350, 0.350, 0.230, 0.260, 0.310, 0.0015, 0.193, 0.200, 0.167, 0.24],
    'Generalization': [0.000, 0.300, 0.400, 0.000, 0.200, 0.180, 0.140, 0.140, 0.100, 0.190, 0.005, 0.083, 0.084, 0.076, 0.085]
}

# Creating a DataFrame from the data
df = pd.DataFrame(data)

# Function to calculate harmonic mean only if all values are greater than zero
def calculate_harmonic_mean(row):
    if row['Old (Locality)'] > 0 and row['New (Reliability)'] > 0 and row['Generalization'] > 0:
        return hmean([row['Old (Locality)'], row['New (Reliability)'], row['Generalization']])
    else:
        return 0

# Apply the harmonic mean calculation to each row and store it in a new column
df['Harmonic Mean'] = df.apply(calculate_harmonic_mean, axis=1)

# Grouping by 'Samples' and calculating the mean and std for each column
summary_df = df.groupby('Samples').agg(
    {
        'Old (Locality)': ['mean', 'std'],
        'New (Reliability)': ['mean', 'std'],
        'Generalization': ['mean', 'std'],
        'Harmonic Mean': ['mean', 'std']
    }
)

# Display the summary DataFrame
summary_df
```
